```python
clean_data = data.replace("\n", " ").strip()
clean_data
```

```python
# Removing punctuations
import string 
translator = str.maketrans("", "", string.punctuation) # str.maketrans(x,y,z) replace x with y and remove z

clean_text = clean_data.translate(translator)
print(clean_text)
```

#### Chunking the entire data into characters of 800 words

```python
max_char = 800 
overlap = 50 
chunks = []
i = 0
while i < len(clean_text):
    piece = clean_text[i: i+max_char]
    chunks.append(piece)
    i = i+max_char - overlap
```

```python
chunks
```

```python
len(chunks)
```

### Using Qwen model to embedd the data

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer("Qwen/Qwen3-Embedding-0.6B")
for i in chunks: 
    embeddings = model.encode(i)
    print(embeddings)
```

```python
emb_list = []
meta = []
for idx, chunk in enumerate(chunks):
    embedding = model.encode(chunk)
    emb_list.append(embedding.astype("float32"))
    meta.append({"chunk": idx, "text": chunk}) 
```

```python
emb_list;
```

```python
meta;
```

```python
import numpy as np 
xb = np.vstack(emb_list)
```

```python
import faiss
faiss.normalize_L2(xb)
```

```python
dim = xb.shape[1]
```

```python
index = faiss.IndexFlatIP(dim)
index.add(xb) # Data is stored in FAISS. It is available in Memory / RAM 
```

```python
index_path = 'index_ramanujan.faiss'
meta_path = 'meta_ramanujan.jsonl' # it is jsonl
```

```python
faiss.write_index(index, index_path)
```

```python
import json, os
with open(meta_path, "w") as f: 
    for item in meta: 
        f.write(json.dumps(item) + "\n")
```
